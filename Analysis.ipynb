```python
path = r'/home/sourav/DataSet/IndianNSEEquity'
files = glob.glob(path + "/*.csv")

print(len(files))

Symbols = []

for file in files:
    Symbols.append(file.split('/')[-1].split('.')[0])
```

```python
testing_time = datetime(2017,1,1)
starting_period = '2017-01-01'
ending_period = '2017-12-31'
```

```python
SelectedFiles = []
SelectedSymbols = []
NoDataCount = 0
Count = 0
for file in files:
    Symbol = Symbols[Count]
    df = pd.read_csv(file)
    df.date = pd.to_datetime(df.date)
    df.set_index('date' , inplace=True)
    
    if df.index[0] > testing_time:
        NoDataCount+=1
        Count+=1
        continue
    
    
    df2017 = df[(df.index > starting_period) & ((df.index < ending_period) )]
    df2017Daily = df2017.resample('D' ).agg({
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last',
        'volume': 'sum',
    }).dropna()
    #SelectedFiles.append(file)
    Volume2017 = df2017Daily['volume'].mean()
    
    if Volume2017 > 100000:
        SelectedFiles.append(file)
        SelectedSymbols.append(Symbol)
    else:
        pass
    

    
    
    Count+=1
    

```

```python
def get_top_stocks(TargetDate , LookBack , NoOfStocks , files,Symbols):
    count = 0
    returndf = pd.DataFrame(columns=['stock','returns', 'path'])
    for file in files:
        df = pd.read_csv(file)
        df.date = pd.to_datetime(df.date)
        df.set_index('date' , inplace=True)
        if df.index[0] > TargetDate:
            continue
        df['1HourCandleReturn'] = ((df.close - df.open)/df.open) * 100
        returns = df.loc[:TargetDate].iloc[-(LookBack+1):-1]['1HourCandleReturn'].sum()
        
        Symbol = Symbols[count]
        
        returndf.loc[len(returndf)] = [Symbol , returns, file]
        
        count+=1
    returndf.sort_values(by='returns', ascending=False , inplace=True)
    returndf.reset_index(drop=True, inplace=True)
    return returndf
        
        
        
    
```
